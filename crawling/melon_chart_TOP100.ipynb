# Melon Top100
- 날짜, 시간 정보, 타이틀, 가수명 수집
- 가수명은 타이틀과 달리, 가수명이 2번 나올 수 있음
- HTML구조를 파악해서 가수명을 1개만 출력하도록 접근
- 해당 경우는, 상위 태그로 구분해서 가져와서 해결함.

```python
from bs4 import BeautifulSoup as bs
import requests as req
import pandas as pd
from IPython.display import display, HTML

#1.
U_A = {'User-Agent':"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/142.0.0.0 Safari/537.36"}
melon_res = req.get("https://www.melon.com/chart/index.htm", headers=U_A)


#2. 파싱
melon_soup = bs(melon_res.text, "lxml")

#3. 요소 
 #1) 날짜와 시간
date = melon_soup.select_one(".calendar_prid.mt12 .year").text
time = melon_soup.select_one(".calendar_prid.mt12 .hour").text

 #2) 타이틀, 가수, 
titles = melon_soup.select(".wrap_song_info .ellipsis.rank01")
titles = [i.text.strip() for i in titles]

singers = melon_soup.select(".ellipsis.rank02 .checkEllipsis")
singers = [i.text.strip() for i in singers]



melon_df = pd.DataFrame({"타이틀":titles,
                        "가수명":singers},index=[str(i)+"위" for i in range(1, len(titles)+1)])
melon_df.index.name="순위"
pd.set_option('display.max_rows', None)
display(HTML('<h2 style="text-align:center">🍈Melon TOP100👻</h2>'))
display(HTML(f'<p style="text-align:center">{date}&nbsp;&nbsp;{time}</p>'))

display(melon_df)
```
